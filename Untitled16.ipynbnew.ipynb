```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/Momen2410/Machine-Learning-Course/main/Data/smoke_detection_iot.csv")
smoke_column = data["Fire_Alarm"]
count_ones = smoke_column[smoke_column == 1].count()
count_zeroes = smoke_column[smoke_column == 0].count()
total_count = count_ones + count_zeroes

percentage_ones = (count_ones / total_count) * 100
percentage_zeroes = (count_zeroes / total_count) * 100

print("Number of ones:", count_ones)
print("Number of zeroes:", count_zeroes)
print("Percentage of ones: {:.2f}%".format(percentage_ones))
print("Percentage of zeroes: {:.2f}%".format(percentage_zeroes))
print("--------------------------------------------------")

values = [percentage_ones, percentage_zeroes]  # Example data
labels = ['Fire', 'No Fire']  # Example labels

plt.pie(values, labels=None, autopct='%1.2f%%')  # Adjusted autopct to show 2 decimal places
plt.title('Fire Alarm Data')
plt.legend(labels, loc='upper right')  # Adjusted the legend position

plt.show()
```
